**Note:** Make sure to have the `concrete_crack_images_for_classification.zip` in the root directory.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import zipfile
from pathlib import Path
```

## 2) Dataset download & preparation

Use `skillsnetwork.prepare()` to download and extract the `concrete_crack_images_for_classification.zip`. 

```python
zip_path = Path("concrete_crack_images_for_classification.zip")

extract_dir = Path("concrete_crack_images_for_classification")
extract_dir.mkdir(exist_ok=True)

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_dir)
```

**Important Note**: 

There are thousands and thousands of images in each folder, so please don't attempt to double click on the folders. This may consume all of your memory and you may end up with a **50*** error. So please **DO NOT DO IT**.   


## 3) Create sorted file-path lists for Negative and Positive folders

Use `os.listdir()` + `sort()` to build consistent, sorted lists of file paths for use in training/validation pipelines.

### Negative directory

```python
negative_images = os.listdir('./concrete_crack_images_for_classification/Negative')
```

```python
negative_images.sort()
```

```python
image_data_negative = Image.open('./concrete_crack_images_for_classification/Negative/{}'.format(negative_images[0]))
image_data_negative
```

Loop through all the images in the <code>./Negative</code> directory and add save their paths.


```python
negative_images_dir = ['./concrete_crack_images_for_classification/Negative/{}'.format(image) for image in negative_images]
```

```python
len(negative_images_dir)
```

### Positive directory

```python
positive_images = os.listdir('./concrete_crack_images_for_classification/Positive')
```

```python
positive_images.sort()
```

```python
image_data_positive = Image.open('./concrete_crack_images_for_classification/Positive/{}'.format(negative_images[0]))
image_data_positive
```

Loop through all the images in the <code>./Positive</code> directory and add save their paths.


```python
positive_images_dir = ['./concrete_crack_images_for_classification/Positive/{}'.format(image) for image in positive_images]
```

```python
len(positive_images_dir)
```
